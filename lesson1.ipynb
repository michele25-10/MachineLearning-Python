```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
X = np.random.standard_normal((200, 1))
y = 2*X + 3
```

```python
plt.plot(X, y)
plt.grid(True)
plt.show()
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])  #input_shape sono il numero di colonne che gli do in input
])
model.summary() #riassunto del modello --> f(x *w  + b)
```

```python
model.compile(loss=tf.keras.losses.MeanSquaredError, optimizer=tf.keras.optimizers.Adam())
history = model.fit(X, y, epochs=1000)
```

```python
y_pred = model.predict(X)
```

```python
loss_hist = history.history['loss']
plt.plot(range(1000), loss_hist)
```
